```python
# Reading the data
data = pd.read_csv("https://osf.io/download/4ay9x/")

# Checking if the data was correctly loaded
data.head()
```

```python
# Filtering for the following occupations:
# Human resources managers (0136)
# Human resources workers (0630)
# Human resources assistants, except payroll and timekeeping (5360)

filtered_data = data[data['occ2012'].isin([136, 630, 5360])]
filtered_data
```

### Create variables

- Female
- Hourly wages = weekly wages/weekly hours
- Log of hourly wages (dependent variable) - We take the log of wages because we are interested in relative differences instead of absolute differences. Relative differences are potentially easier to interpret if we are talking about wage differences.
- Level of education

```python
# Creating the variables detailed in the markdown cell
filtered_data["female"] = (filtered_data["sex"] == 2)
filtered_data["w"] = filtered_data["earnwke"] / filtered_data["uhours"]
filtered_data["lnw"] = np.log(filtered_data["w"])
filtered_data["edu"] = filtered_data["grade92"]

# Checking the new variables
filtered_data
```

## Distribution of earnings

```python
# Checking for log distribution

(
    ggplot(filtered_data, aes(x="lnw", y="stat(count)/sum(stat(count))"))
    + geom_histogram(
        binwidth=0.1,
        boundary=0,
        color="white",
        fill = "darkblue",
        size=0.25,
        alpha=0.8,
        show_legend=False,
        na_rm=True,
    )
    + labs(x="Log of hourly wages", y="Percent")
    + expand_limits(x=0.01, y=0.01)
    + scale_x_continuous(expand=(0.01, 0.01), limits=(0, 5))
    + scale_y_continuous(
        labels=percent_format(), #mizani
        limits=(0, 0.1),
        expand=(0.0, 0.0),
    )
    + theme_bw()
)

# The distribution of ln wages is close to normal
```

```python
# Checking mean values in the data set
filtered_data.groupby("female").mean()
```

```python
# Creating the regression and checking the summary of regression
reg1 = smf.ols(formula="lnw~female", data=filtered_data).fit(cov_type="HC1")
reg1.summary()
```

### Intepreting our key coefficients:

- (1) Slope: female HR employees earn 15.7% less, on average than their male counterparts in this data.

- (2) R squared is 0.018 which means that gender explains very little (less than 2%) of the variation in earnings in this data.

- (3) The confidence interval of the slope is -0.1571+1.96*0.038=-0.083; -0.1571-1.96*0.038=-0.232 ==> [-0.232;-0.083] which does not include zero. This means that wwe can be 95% confident that the average earning difference between males and females HR employees is between -23.2% and -0.83%. 

- (4) This means that we can still rule out with a 95\% confidence that their average earnings (female and male) are the same.

- (5) The p-value < 0.05 also proves the above statement.

- (6) Note: As p-value < 0.01, the above statement still holds with 99% confidence. 

# Level of education

```python
# Checking the amount of observations in each level of education category
filtered_data.grade92.value_counts().sort_values
```

```python
filtered_data
```

```python
# Filtering for categories 39-46
filtered_data = filtered_data[filtered_data["grade92"] >= 39]
filtered_data

# Note: 10 observations were dropped
```

```python
# Grouping educational levels and creating dummies
filtered_data["ed_HS"] = (filtered_data["grade92"] == 39).astype(int)
filtered_data["ed_college"] = (filtered_data["grade92"].isin([40, 41 ,42 , 43]).astype(int))
filtered_data["ed_MA_Prof"] = (filtered_data["grade92"].isin([44, 45]).astype(int))
filtered_data["ed_Phd"] = (filtered_data["grade92"] == 46).astype(int)
filtered_data
```

```python
# In reg3, category "ed_HS" is our reference group, hence we are not taking it into consideration
# In reg4, category "ed_MA_Prof" is our reference group, hence we are not taking it into consideration

reg2 = smf.ols(formula="lnw~female", data=filtered_data).fit(cov_type="HC1")
reg3 = smf.ols(formula="lnw~female + ed_college + ed_MA_Prof + ed_Phd", data=filtered_data).fit(cov_type="HC1")
reg4 = smf.ols(formula="lnw~female + ed_HS + ed_college + ed_Phd", data=filtered_data).fit(cov_type="HC1")
```

```python
# Checking the regression comparison table with Stargazer
stargazer = Stargazer([reg2, reg3, reg4])
stargazer.covariate_order(["female[T.True]", "ed_HS", "ed_college", "ed_MA_Prof", "ed_Phd", "Intercept"])
stargazer.rename_covariates({"female[T.True]": "female", "Intercept": "Constant"})
stargazer
```

## Interpreting coefficients in the *second column* about HR employees

In this regression, ed_HS was used as the comparison group.

- Female HR employees earn 14.0% less, on average than their male counterparts with identical level of education.

Employees of the same gender with:

- (1) college degrees (some college experience, associate degrees, BA degrees) expected to earn 13.7% more;
- (2) MA or Professional degrees (the category with the most observations) expected to earn 31.6% more;
- (3) Phd degrees expected to earn 67.2% more

than people only holding a high school diploma.

- (4) All covariates are significant at 99%.

- (5) R-squared is 0.054 which means the model explains 5.4% of the variation in ln wages.
- (6) R-squared is higher than R-squared only considering gender (0.017). This means that by taking level of education into consideration, we increase our model's goodness of fit. 
- (7) In the data, women appear to be more likely to be in lower-earning categories than in higher-earner - but only small part since the female coefficient for column 2 is lower in absolute value than for column 1 (-0.14 vs -0.154).

## Interpreting coefficients in the *third column* about HR employees
